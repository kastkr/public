```python
import pandas as pd
import numpy as np
```

```python
#create variable for wiki url
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
#use pandas library to read the table into a dataframe
df = pd.read_html(url)[0]
df
```

```python
#correct column headers
df.columns = df.iloc[0]
df = df[1:]
df
```

```python
#drop rows where Borough is "Not assigned"
df = df[df.Borough != 'Not assigned']
df
```

```python
#Group by Postcode and join Neighbourhoods
df = df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df
```

```python
#Identify Neighbourhood's that are not assigned
df.loc[df['Neighbourhood'] == 'Not assigned']
```

```python
#Replace the Not assigned neighborhood w/ the Borough name and then check work
df.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df['Borough']
df.loc[df['Postcode'] == 'M7A']
```

```python
#show the shape of the final dataframe
df.shape
```

```python
#geocoder would not work for me - so leveraging the provided file
#pulling into a net new dataframe that can be used to join on Postal Code / Postcode
lat_long = pd.read_csv('http://cocl.us/Geospatial_data')
lat_long
```

```python
#merge the two tables using the match on Postcode
df = df.merge(lat_long, left_on='Postcode', right_on='Postal Code')
```

```python
#drop the extra column
df = df.drop(columns = ['Postal Code'])
```

```python
df
```

```python
#import modules to do clustering and visualization
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
from sklearn.cluster import KMeans
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
address = 'Toronto, CAN'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df['Latitude'], df['Longitude'], df['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
# set number of clusters
kclusters = 4

# create data set with just lat/long features
df_clustering = df[['Latitude', 'Longitude']]

# run k-means clustering to cluster based off of geographical positioning
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
df.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df.head()
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df['Latitude'], df['Longitude'], df['Neighbourhood'], df['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
